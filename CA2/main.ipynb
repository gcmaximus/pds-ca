```python
# Author: Jerald Yeo
# Title: Statistics on Health Behaviours among Secondary School Students from 2006 to 2012
# Dataset 1: https://data.gov.sg/dataset/students-health-survey?resource_id=8797f146-aee6-4942-928d-b00af556b624

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl


#READ DATA
df = pd.read_csv("data/health-behaviours-among-secondary-school-students-by-educational-level.csv",
                   sep=",", index_col='year')
df = df.replace('na',float('NaN'))

#NATURE OF DATA
print('=========================================================')
print('#   Health Behaviours among Secondary School Students   #')
print('#                 from 2006 to 2012                     #')
print('=========================================================')
print()
print('[ Question ]')
print('------------')
print('How do the different health behaviours among secondary school students vary?')
print()
print('[ Nature of the dataset ]')
print('-------------------------')
print('The dataset contains data on the presence of different health behaviours in')
print('secondary school students in 2006, 2009 and 2012.')
print('**********************************************************************************')
print(f'The dataset contains {df.shape[0]} rows and {df.shape[1]} columns.')
print('**********************************************************************************')
print(f'INDEX of dataset: \n{df.index}\n')
print('**********************************************************************************')
print(f'COLUMNS of dataset: \n{df.columns}\n')
print('**********************************************************************************')
print(f'NUMBER OF NON-NA VALUES in dataset:')
print(f'{df.count()}\n')
print('**********************************************************************************')
print(f'MORE GENERAL INFORMATION on dataset:')
print(f'{df.info()}\n')
print('**********************************************************************************')
print(f'DETAILED STATISTICAL SUMMARY of dataset:')
print(f'{df.describe()}\n')
print('**********************************************************************************')

#DATA VISUALISATION
print('[ Data Visualisation ]')
print('----------------------')

#PREPARING THE DATA

#filtering out data only with "Overall" education level
df = df.iloc[[0,3,6]]
#removing "edu_level" column from dataframe
df = df.drop('edu_level', axis=1)
#remove columns with na values (fat_intake and binge_drinking)
df = df.dropna(axis=1)
#convert to float values
df = df.astype(float)


#CREATING 3 SUBPLOTS
fig, axs = plt.subplots(1,3, figsize=(13,5))
plt.suptitle('Health behaviours among Secondary School Students from 2006 to 2012')
mpl.style.use('bmh')

#SUBPLOT 1,2,3: Bar graphs
labels = ('Physical activity', 'Fruits', 'Vegetables', 'Fruits & Vegetables',
          'Sweetened Drinks', 'Deep Fried Food', 'Salt use', 'Smoking')
years = df.index.unique()
colors = ('blue','red','purple')

#creating 3 separate charts for each year
for num in range(3):
    
    #setting up the graph
    year = years[num]
    axs[num].set_title(str(year))
    axs[num].set_xlabel('Health Behaviours')
    axs[num].set_ylabel('Percentage of students (%)')
    axs[num].set_xticks([i for i in range(len(labels))])
    axs[num].set_xticklabels(labels=labels, rotation=90)
    
    #obtaining values for each year
    values = df.loc[year]
    
    #plotting the graph
    axs[num].bar(labels, values, color=colors[num])

#SHOW THE GRAPH
plt.show()


#INSIGHTS
print('[ Insights! ]')
print('-------------')
print('From the bar graphs:')
print()
print('1. In 2006, 2009 and 2012, the percentages of students possessing the different health behaviours stay relatively the same.')
print('2. The largest change in percentage of health behaviours is the decrease in consumption of Fruits from 2006 to 2009') 
print('\t(~12% decrease).')
print("3. The students' consumption of Sweetened Drinks was always the most common out of the health behaviours (~75%).")
print('4. Meanwhile, Smoking was always the least common out of the health behaviours (~7%).')
```

```python
# Author: Jerald Yeo
# Title: Statistics on Vaccination and Immunisation of students
# Dataset 2: https://data.gov.sg/dataset/vaccination-and-immunisation-of-students?view_id=5535aaa1-bc9b-4daf-b806-c74adcc7b278&resource_id=fbd97640-bf90-43ad-8dba-37d0a25e8658

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

#READ DATA
df = pd.read_csv("data/vaccination-and-immunisation-of-students-annual.csv",
                   sep=",", index_col='year')

#NATURE OF DATA
print('================================================')
print('#   Vaccination and Immunisation of students   #')
print('#            from 2009 to 2019                 #')
print('================================================')
print()
print('[ Question ]')
print('------------')
print('1) How does the 3 different types of vaccinations taken by students vary in number throughout the years?')
print('2) What are the distributions of the different vaccinations taken by students?')
print()
print('[ Nature of the dataset ]')
print('-------------------------')
print('The dataset contains data on the different number of doses of different vaccines taken by students from 2009 to 2019.')
print('**********************************************************************************')
print(f'The dataset contains {df.shape[0]} rows and {df.shape[1]} columns.')
print('**********************************************************************************')
print(f'INDEX of dataset: \n{df.index}\n')
print('**********************************************************************************')
print(f'COLUMNS of dataset: \n{df.columns}\n')
print('**********************************************************************************')
print(f'NUMBER OF NON-NA VALUES in dataset:')
print(f'{df.count()}\n')
print('**********************************************************************************')
print(f'MORE GENERAL INFORMATION on dataset:')
print(f'{df.info()}\n')
print('**********************************************************************************')
print(f'DETAILED STATISTICAL SUMMARY of dataset:')
print(f'{df.describe()}\n')
print('**********************************************************************************')

#DATA VISUALISATION
print('[ Data Visualisation ]')
print('----------------------')


#CREATING 2 SUBPLOTS
fig, axs = plt.subplots(2,1, figsize=(13,13))
plt.suptitle('No. of Vaccination and Immunisation of students from 2009 to 2019')
mpl.style.use('bmh')

#SUBPLOT 1 - Line chart
#setting up the graph
labels = [i for i in range(2009,2020)]
axs[0].set_title('Change in No. of students taking different vaccines over the years')
axs[0].set_xlabel('Years')
axs[0].set_ylabel('No. of students vaccinated (Thousands)')
axs[0].set_xticks(labels)
axs[0].set_xticklabels(labels=labels)
markers = ('^','o','*')
vaccines = df['vaccination_type'].unique()

#creating one line for each vaccine type
for i in range(3):
    
    #obtaining values for each vaccine
    values = df[df['vaccination_type'] == vaccines[i]]['no_of_doses_in_thousands']
    
    #plotting the lines
    axs[0].plot(labels, values, marker=markers[i], label=vaccines[i].replace(' ', ', '))


#show the legend
leg = axs[0].legend()
leg.set_title('Vaccination Types', prop={
    'size': 12,
    'weight': 'bold'
})

#SUBPLOT 2 - Boxplot
#setting up the graph
labels = df['vaccination_type'].unique()
axs[1].set_title('Distribution in vaccines taken by students')
axs[1].set_xlabel('Vaccines')
axs[1].set_ylabel('No. of students vaccinated in a year (Thousands)')
axs[1].set_xticks([i for i in range(1,len(labels)+1)])
axs[1].set_xticklabels(labels=labels)

#obtaining values for each vaccine
values = []
for i in range(3):
    
    #obtaining values for each vaccine
    vaccine_values = df[df['vaccination_type'] == labels[i]]['no_of_doses_in_thousands']
    values.append(vaccine_values)
    
#plotting the graph
labels = [label.replace(' ',', ') for label in labels]
axs[1].boxplot(values, labels=labels)


#SHOW THE GRAPH
plt.show()



#INSIGHTS
print('[ Insights! ]')
print('-------------')
print('From the line graph:')
print()
print('1. Diphtheria, Tetanus vaccine remains consistent in the number of doses over the years (~40-50k).')
print('2. Poliomyelitis and Measles, Mumps, Rubella vaccines decrease generally in number of doses over the years.')
print('3. Measles, Mumps, Rubella vaccine has the largest drop in number of doses from 2009 to 2019 (~85k decrease).')
print('4. There was a sudden drastic drop in Measles, Mumps, Rubella vaccine doses, from 2011 to 2012.')
print('5. There was also a sudden drastic drop in Poliomyelitis vaccine doses, from 2013 to 2014.')
print()
print('From the box plot:')
print()
print('1. Measles, Mumps and Rubella vaccine has the highest spread in doses.')
print('2. Diphtheria, Tetanus vaccine vaccine has the lowest spread in doses.')
print('3. The largest number of students dosed by a vaccine in a year is by the Poliomyelitis vaccine (~90k).')
print('4. The smallest number of students dosed by a vaccine in a year is by the Measles, Mumps, Rubella vaccine (~5k).')


```

```python
# Author: Jerald Yeo
# Title: Statistics on defective vision of students examined
# Dataset 3: https://data.gov.sg/dataset/common-health-problems-of-students-examined-obesity-annual?view_id=416d850a-152b-4037-b437-61659f81dbef&resource_id=7f8abd13-5662-46ac-9c9b-12718ce973ea

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

#READ DATA
df = pd.read_csv("data/common-health-problems-of-students-examined-overweight-annual.csv",
                   sep=",", index_col='year')

#NATURE OF DATA
print('============================================')
print('#   Overweight rate of students examined   #')
print('#            from 2009 to 2020             #')
print('============================================')
print()
print('[ Question ]')
print('------------')
print('How does the number of overweight students differ in terms of gender and age group?')
print()
print('[ Nature of the dataset ]')
print('-------------------------')
print('The dataset contains data on the number of overweight students per 10000 students examined')
print('in different age groups and genders, from 2009 to 2020.')
print('**********************************************************************************')
print(f'The dataset contains {df.shape[0]} rows and {df.shape[1]} columns.')
print('**********************************************************************************')
print(f'INDEX of dataset: \n{df.index}\n')
print('**********************************************************************************')
print(f'COLUMNS of dataset: \n{df.columns}\n')
print('**********************************************************************************')
print(f'NUMBER OF NON-NA VALUES in dataset:')
print(f'{df.count()}\n')
print('**********************************************************************************')
print(f'MORE GENERAL INFORMATION on dataset:')
print(f'{df.info()}\n')
print('**********************************************************************************')
print(f'DETAILED STATISTICAL SUMMARY of dataset:')
print(f'{df.describe()}\n')
print('**********************************************************************************')

#DATA VISUALISATION
print('[ Data Visualisation ]')
print('----------------------')


#CREATING 2 SUBPLOTS
fig, axs = plt.subplots(1,2, figsize=(13,7))
plt.suptitle('Pie charts of Overall Overweight Primary School Students from 2009 to 2020')
mpl.style.use('bmh')

#SUBPLOT 1: Pie Chart of Genders
#setting up the graph
gender_labels = df['gender'].unique()
explode = (0, 0.1)
colors = ('#4fa7ff','#ff4de7')
axs[0].set_title('Gender of all overweight students')

#obtaining values for each gender
gender_values = []
for gender in gender_labels:
    
    #calculating the sum of all values across all years of different genders
    gender_values.append(df[df['gender'] == gender]['per_10000_examined'].sum())
    
#plotting the graph
axs[0].pie(gender_values, labels=gender_labels, autopct='%1.1f%%', 
          shadow=True, explode=explode, colors=colors)

#show the legend
axs[0].legend()



#SUBPLOT 2: Pie Chart of Age Group
#setting up the graph
age_group_labels = df['age_group'].unique()
explode = (0, 0.1)
colors = ('#4ef51b','#f58b1b')
axs[1].set_title('Age Group of all overweight students')

#obtaining values for each age group
age_group_values = []
for age_group in age_group_labels:
    
    #calculating the sum of all values across all years of different age groups
    age_group_values.append(df[df['age_group'] == age_group]['per_10000_examined'].sum())
    
#plotting the graph
axs[1].pie(age_group_values, labels=age_group_labels, autopct='%1.1f%%', 
          shadow=True, explode=explode, colors=colors, startangle=180)

#show the legend
axs[1].legend(loc='lower right')



#SHOW THE GRAPH
plt.show()



#INSIGHTS
print('[ Insights! ]')
print('-------------')
print('From the pie charts:')
print()
print('1. Male students had more overweight cases than female students throughout the years.')
print('2. Primary 5 students had more overweight cases than Primary 1 students throughout the years.')


```

```python
# Author: Jerald Yeo
# Title: Statistics on Weekly Number of Dengue and Dengue Haemorrhagic Fever Cases
# Dataset 4: https://data.gov.sg/dataset/weekly-number-of-dengue-and-dengue-haemorrhagic-fever-cases?view_id=40a6c1fa-842c-420e-ab09-1676a24b2729&resource_id=c6db5eb2-6daf-4d3b-b571-7bc16c8f03c4

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

#READ DATA
df = pd.read_csv("data/weekly-number-of-dengue-and-dengue-haemorrhagic-fever-cases.csv",
                   sep=",", index_col='year')

#NATURE OF DATA
print('=====================================')
print('#   Weekly number of Dengue Cases   #')
print('#        from 2014 to 2018          #')
print('=====================================')
print()
print('[ Question ]')
print('------------')
print('What is the correlation between the number of weeks passed and the number of weekly Dengue cases?')
print()
print('[ Nature of the dataset ]')
print('-------------------------')
print('The dataset contains data on weekly number of Dengue and DHF cases from 2014 to 2018.')
print('**********************************************************************************')
print(f'The dataset contains {df.shape[0]} rows and {df.shape[1]} columns.')
print('**********************************************************************************')
print(f'INDEX of dataset: \n{df.index}\n')
print('**********************************************************************************')
print(f'COLUMNS of dataset: \n{df.columns}\n')
print('**********************************************************************************')
print(f'NUMBER OF NON-NA VALUES in dataset:')
print(f'{df.count()}\n')
print('**********************************************************************************')
print(f'MORE GENERAL INFORMATION on dataset:')
print(f'{df.info()}\n')
print('**********************************************************************************')
print(f'DETAILED STATISTICAL SUMMARY of dataset:')
print(f'{df.describe()}\n')
print('**********************************************************************************')

#DATA VISUALISATION
print('[ Data Visualisation ]')
print('----------------------')

#PREPARING THE DATA
#select only the records where type_dengue == dengue
df = df[df['type_dengue'] == 'Dengue']
original_len = len(df) #needed for xticks later
#remove rows with NA values
df = df.dropna()
#merging the year and week column into year-week column
df['year-week'] = df.apply(lambda x: str(x.name) + '-' + str(x['eweek']), axis=1)
#reset the index column to "year-week" column
df = df.set_index(['year-week'])


#CREATING 1 PLOT
plt.figure(figsize=(13,7))
plt.title('Correlation between Weeks passed and No. of Weekly Dengue Cases from 2014 to 2018')
mpl.style.use('bmh')


#PLOT: Scatter plot
#setting up the graph
plt.xlabel('Weeks')
plt.ylabel('No. of Cases in Week')
labels = [i for i in range(2014,2019)]
x_ticks = [i for i in range(0,original_len,53)] # 5 xticks, spread & located at the start of each year (every 53 weeks)
plt.xticks(x_ticks, labels=labels, rotation=25)
weeks = df.index.unique()

#obtaining values
values = df['number']
    
#plotting the graph
plt.scatter(weeks, values, color='#e35f5f')

#SHOW THE GRAPH
plt.show()



#INSIGHTS
print('[ Insights! ]')
print('-------------')
print('From the scatter plot:')
print()
print('1. From 2014 to 2017, the number of dengue cases in a week is more spread out and fluctuates more (range of ~100 to ~900)')
print('compared to 2017 to 2018, which number of dengue cases in a week is more consistent (range of ~40 to ~180).')
print('2. There are 2 times when the number of Dengue cases in a week peaked suddenly:')
print('\t- Middle of 2014 (~900)')
print('\t- Start of 2016 (~650)')
print('3. Generally, the number of weekly cases from 2014 to 2017 is much higher than the number from 2017 to 2018.')
print('4. There are little to no anomalies in the data.')
```
