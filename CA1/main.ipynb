```python
# Author: Jerald Yeo
# Title: Statistics on Total Air Passenger Departures from 2013-01 to 2020-01
# Dataset 1: https://data.gov.sg/dataset/air-passenger-departures-total-by-region-and-selected-country-of-disembarkation?resource_id=04e41bb7-ad8d-4a25-b036-7e4057b074a8
# Question: What is the distribution in the different regions in the world
#           that air passengers departing from Singapore are disembarking at?
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

#READ DATA
data = np.genfromtxt('data/total-air-passenger-departures-by-region.csv',
                    delimiter=',',
                    names=True, dtype=('U7','U34','U15',int))

#NATURE OF DATA
print('=======================================================================')
print('=== << Air Passenger Departures from Singapore from 2013 to 2019 >> ===')
print('=======================================================================')
print()

print("[ What's in this dataset? ]")
print('---------------------------')
months = np.unique(data["month"])
regions = np.unique(data["level_2"])
print('The dataset contains data on how many passengers from Singapore disembark at different regions in the world every month.')
print()
print(f'There are {len(data)} rows in the dataset')
print(f'There are {len(months)} months of data from {months[0]} to {months[-1]}')
print(f'There are {len(regions)} different regions passengers disembark at')
print()

print('[ Data Visualization ]')
print('----------------------')
print('Note: Data from 2013 to 2019 only will be extracted from the dataset to be analysed.')
print('This is because some regions from 1961 to 2012 have NA values which we cannot compare.')
print()

#filtering out data only from 2013 to 2019
boolean = data["month"] >= "2013-01"
data_2013_2019 = data[boolean]
boolean = data_2013_2019["month"] <= "2019-12"
data_2013_2019 = data_2013_2019[boolean]
data = data_2013_2019

#obtaining new months list from 2013 to 2019
months = np.unique(data["month"])

#CREATING 2 SUBPLOTS
fig, axs = plt.subplots(1,2, figsize=(13,7))
plt.suptitle(f'Air Passenger Departures from Singapore from {months[0][:4]} to {months[-1][:4]}')
mpl.style.use('ggplot')
axs[0].ticklabel_format(useOffset=False, style='plain')

#SUBPLOT 1: BAR GRAPH - No. of Air Passenger Departures by region
axs[0].set_title('No. of Air Passenger Departures by region')
axs[0].set_xlabel('Region of Disembarkation')
axs[0].set_ylabel('No. of Air Passengers')


#set up y-values
y_values = []
for region in regions:
    boolean = data["level_2"] == region
    values = data[boolean]["value"]
    sum = values.sum()
    y_values.append(sum)
    
#convert to numpy array
y_values = np.array(y_values)

#set up x-values
x_ticks = np.arange(len(regions))
x_values = regions
axs[0].set_xticks(x_ticks, labels=regions, rotation=90)

#plot the graph
axs[0].bar(x_values, y_values)

#SUBPLOT 2: PIE CHART - Composition of Air Passenger Departures by region
axs[1].set_title('Composition of Air Passenger Departures by region')

#setting up
explode = (0,0,0,0,0,0,0,0.1)
colors = ('pink','gray','yellow','green','olive','red','orange','cyan')

#plot the graph
axs[1].pie(y_values,explode=explode,
           labels=regions,autopct='%1.1f%%',
          shadow=True, colors=colors)

plt.legend(loc='lower left')

#TEXT ANALYSIS
for i in range(len(regions)):
    print(f'<< Region of disembarkation: {regions[i]} >>')
    print('=================================================')
    print(f'Total number of air passengers: {y_values[i]:,}.')
    print()

min_passengers = y_values.min()
argmin_passengers = y_values.argmin()
max_passengers = y_values.max()
argmax_passengers = y_values.argmax()
mean_passengers = math.floor(y_values.mean())
sum_passengers = y_values.sum()
print(f'The region with the least number of air passengers departing to is {regions[argmin_passengers]}, with a total of {min_passengers:,} passengers.')
print(f'The region with the most number of air passengers departing to is {regions[argmax_passengers]}, with a total of {max_passengers:,} passengers.')
print(f'The mean number of air passengers departed from Singapore is around {mean_passengers:,} passengers.')
print(f'The total number of air passengers departed from Singapore is {sum_passengers:,} passengers.')
print()

#SHOW THE GRAPH
plt.show()


#INSIGHTS
print('[ Insights! ]')
print('-------------')
print(f'{regions[argmax_passengers]} is the region with the most number of passengers departing to.')
print(f'{regions[argmin_passengers]} is the region with the least number of passengers departing to.')
```

```python
# Author: Jerald Yeo
# Title: Statistics on Visitor International Arrivals to Singapore
# Dataset 2: https://data.gov.sg/dataset/total-visitor-international-arrivals-to-singapore?view_id=4723beba-7ec9-4357-bfea-27d13aa06729&resource_id=f8c014e4-fc08-4e28-a1b8-27a8390afd1e
# Question: How does the number of visitor international arrivals in Singapore change as time passes?
#           What is the frequency distribution of visitor international arrivals in Singapore?
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

#READ DATA
data = np.genfromtxt('data/total-visitor-international-arrivals-monthly.csv',
                    delimiter=',',
                    names=True, dtype=[('month', 'U7'), ('total_international_visitor_arrivals', int)])

#NATURE OF DATA
print('===========================================================================')
print('=== << Visitor International Arrivals to Singapore from 1978 to 2014 >> ===')
print('===========================================================================')
print()

print("[ What's in this dataset? ]")
print('---------------------------')
months = np.unique(data["month"])
print('The dataset contains data on how many international visitor arrivals Singapore receives every month.')
print()
print(f'There are {len(data)} rows in the dataset')
print(f'There are {len(months)} months of data from {months[0]} to {months[-1]}')
print()


print('[ Data Visualization ]')
print('----------------------')
print('Note: Data until 2014 only will be extracted from the dataset to be analysed for easier comparison between the years.')
print()

#filtering out data until 2014-12
boolean = data["month"] <= "2014-12"
data_2015_01 = data[boolean]
data = data_2015_01

#obtaining new months list until 2014-12
months = np.unique(data["month"])

#CREATING 2 SUBPLOTS
fig, axs = plt.subplots(2,1, figsize=(13,11))
plt.suptitle(f'Total Visitor International Arrivals to Singapore from {months[0][:4]} to {months[-1][:4]}')
plt.tight_layout(pad=5)
mpl.style.use('ggplot')
axs[0].ticklabel_format(useOffset=False, style='plain')
axs[1].ticklabel_format(useOffset=False, style='plain')

#SUBPLOT 1: LINE GRAPH - Total Monthly Visitor International Arrivals to Singapore
axs[0].set_title(f'Total Monthly Visitor International Arrivals to Singapore')
axs[0].set_xlabel('Time (Months)')
axs[0].set_ylabel('No. of Visitor Arrivals')

#set up x-values
x_labels = np.arange(1978,2015)
x_values = months
x_ticks = np.arange(0,len(months),12)
axs[0].set_xticks(x_ticks, labels=x_labels, rotation=45)

#set up y-values
monthly_visitors = data["total_international_visitor_arrivals"]
y_values = monthly_visitors

#plot the graph
axs[0].plot(x_values, y_values, 'g')
axs[0].annotate('Massive drop in \nvisitors around 2003', xy=('2003-07',180000), xytext=('2006-01',340000),
                bbox={
                    'boxstyle':'round4,pad=.5',
                    'facecolor':'#02bd49',
                    'edgecolor':'green',
                    'alpha':0.8
                },
                arrowprops={
                    'color':'black',
                    'headwidth':9,
                    'width':3
                })

#SUBPLOT 2: HISTOGRAM - Frequency of Monthly Visitor International Arrivals to Singapore
axs[1].set_title('Frequency of Monthly Visitor International Arrivals to Singapore')
axs[1].set_xlabel('No. of Visitor Arrivals')
axs[1].set_ylabel('Frequency')

#plot the graph
axs[1].hist(y_values, bins=60, color='blue',
            alpha=0.9, edgecolor='gray')

#TEXT ANALYSIS
min_visitors = y_values.min()
argmin_visitors = y_values.argmin()
max_visitors = y_values.max()
argmax_visitors = y_values.argmax()
mean_visitors = math.floor(y_values.mean())
sum_visitors = y_values.sum()

print(f'The lowest number of international visitor arrivals to Singapore is {min_visitors:,} visitors, on {data["month"][argmin_visitors]}.')
print(f'The highest number of international visitor arrivals to Singapore is {max_visitors:,} visitors, on {data["month"][argmax_visitors]}.')
print(f'The mean number of international visitor arrivals to Singapore is around {mean_visitors:,} visitors.')
print(f'The total number of international visitor arrivals to Singapore is {sum_visitors:,} visitors.')
print()

#SHOW THE GRAPH
plt.show()


#INSIGHTS
print('[ Insights! ]')
print('-------------')
print('As time passes on, the number of international visitors arrivals to Singapore generally increases.')
print('The rise in visitor arrivals from 1978 - 2004 is much more gradual compared to the rise from 2004 - 2014.')
print('Around the start of 2003, the number of visitors suddenly falls drastically, but rises sharply again between 2003 and 2004.')
print()

print('Most frequently, there are around 100,000 to 300,000 visitor arrivals to Singapore in a month.')
print('Least frequently, there are around 1,000,000 to 1,500,000 visitor arrivals to Singapore in a month.')
```

```python
# Author: Jerald Yeo
# Title: Statistics on weight of air cargo from different countries discharged at Singapore
# Dataset 3: https://data.gov.sg/dataset/air-cargo-discharged-by-region-country-of-origin-monthly?resource_id=afe8d68e-60f8-434c-ab1a-255a67c635a0
# Question: What's the difference in the distribution of the weight of air cargo originating from different countries?
#           How do time and the total weight of air cargo from all countries discharged correlate?
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

#READ DATA
data = np.genfromtxt('data/total-air-cargo-discharged-in-tonnes-by-country.csv',
                    delimiter=',',
                    names=True, dtype=('U7','U5','U15','U14','f4'))

#NATURE OF DATA
print('====================================================')
print('=== << Air Cargo Discharged from 1999 to 2018 >> ===')
print('====================================================')
print()

print("[ What's in this dataset? ]")
print('---------------------------')
months = np.unique(data["month"])
regions = np.unique(data["level_2"])
countries = np.unique(data["level_3"])
print('The dataset contains data on how much air cargo from different countries is discharged at Singapore every month.')
print()
print(f'There are {len(data)} rows in the dataset')
print(f'There are {len(months)} months of data from {months[0]} to {months[-1]}')
print(f'There are {len(regions)} different regions and {len(countries)} countries of origin')
print()

print('[ Data Visualization ]')
print('----------------------')
print('Note: Data from 1999 to 2018 only will be extracted from the dataset to be analysed.')
print('This is because some countries from 1986 to 1998 have NA values which we cannot compare.')
print()

#filtering out data only from 1999 to 2018
boolean = data["month"] >= "1999-01"
data_1999_2018 = data[boolean]
boolean = data_1999_2018["month"] <= "2018-12"
data_1999_2018 = data_1999_2018[boolean]
data = data_1999_2018

#obtaining new months list from 1999 to 2018
months = np.unique(data["month"])

#CREATING 2 SUBPLOTS
fig, axs = plt.subplots(2,1, figsize=(13,15))
plt.suptitle(f'Air Cargo Discharged at Singapore from {months[0][:4]} to {months[-1][:4]}')
mpl.style.use('seaborn-notebook')
plt.tight_layout(pad=7)

#SUBPLOT 1: BOX PLOT - Weight of Air Cargo from different countries
axs[0].set_title('Weight of Air Cargo from different countries')
axs[0].set_xlabel('Country of Origin')
axs[0].set_ylabel('Air Cargo Discharged (Tonnes)')

#set up y-values
y_values = []
for country in countries:
    boolean = data["level_3"] == country
    values = data[boolean]["value"]
    y_values.append(values)

#convert to numpy array
y_values = np.array(y_values)

#set up x-values
x_ticks = np.arange(1,len(countries)+1)
axs[0].set_xticks(x_ticks, labels=countries, rotation=25)

#plot the graph
y_values = y_values.transpose()
axs[0].boxplot(y_values, labels=countries, patch_artist=True,
               boxprops={'facecolor':'#830afc'},
               medianprops={'color':'black'}
)

box_y_values = y_values

#SUBPLOT 2: SCATTER PLOT - Correlation between Time and Total Weight of Air Cargo from all 11 countries
axs[1].set_title(f'Correlation between Time and Total weight of Air Cargo from all {len(countries)} countries')
axs[1].set_xlabel('Time (Months)')
axs[1].set_ylabel('Total Weight of Air Cargo (Tonnes)')

#set up y-values
y_values = []
for month in months:
    boolean = data["month"] == month
    values = data[boolean]["value"]
    y_values.append(values)
    
#convert to numpy array
y_values = np.array(y_values)
y_values = np.sum(y_values, axis=1)

#set up x-values
x_labels = np.arange(1999,2019)
x_values = months
x_ticks = np.arange(0,len(x_values),12)
axs[1].set_xticks(x_ticks, labels=x_labels, rotation=25)

#plot the graph
axs[1].scatter(x_values, y_values, marker='^', color='#0011ff')
axs[1].text('2014-01', 32000, 'Positive Correlation', style='italic',
            bbox={
               'facecolor':'orange',
               'edgecolor':'black',
               'pad':10,   
            })

scatter_y_values = y_values

#TEXT ANALYSIS
for i in range(len(countries)):
    min_weight = box_y_values.transpose()[i].min()
    max_weight = box_y_values.transpose()[i].max()
    median_weight = np.median(box_y_values.transpose()[i])
    std_weight = np.std(box_y_values.transpose()[i])
    print(f'<< Country of origin: {countries[i]} >>')
    print('========================================')
    print(f'Lightest weight of air cargo in one month: {min_weight:.2f} tonnes.')
    print(f'Heaviest weight of air cargo in one month: {max_weight:.2f} tonnes.')
    print(f'Median weight of air cargo: {median_weight:.2f} tonnes.')
    print(f'Standard deviation of air cargo: {std_weight:.2f} tonnes.')
    print()

total_min_weight = scatter_y_values.min()
total_argmin_weight = scatter_y_values.argmin()
total_max_weight = scatter_y_values.max()
total_argmax_weight = scatter_y_values.argmax()

print(f'The lowest total weight of air cargo from all 11 countries is {total_min_weight} tonnes, on {months[total_argmin_weight]}.')
print(f'The highest total weight of air cargo from all 11 countries is {total_max_weight} tonnes, on {months[total_argmax_weight]}.')
print()

#SHOW THE GRAPH
plt.show()



#INSIGHTS
print('[ Insights! ]')
print('-------------')
print('China has the highest spread in air cargo weights.')
print('Germany has the lowest spread in air cargo weights.')
print('The heaviest air cargo discharged originated from China.')
print('The lightest air cargo discharged originated from Vietnam.')
print()

print('There is a positive correlation between months passed and total weight of air cargo discharged.')
```
